```python
import pandas as pd
df= pd.read_csv('data.csv')
```

```python
df
```

```python
def vectorize_input(symptoms_arr):
    symptoms_vector = []
    for column in df.drop('Plant', axis=1).columns:
        if column in symptoms_arr:
            symptoms_vector.append(1)
        else:
            symptoms_vector.append(0)
    return symptoms_vector
```

```python
from sklearn.neighbors import KNeighborsClassifier
X = df.drop('Plant', axis=1)
y = df['Plant']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,train_size=0.2,random_state=42)
```

```python
def train_knn_model(n_neighbors=3):
    knn_model = KNeighborsClassifier(n_neighbors=n_neighbors)
    knn_model.fit(x_train,y_train)
    return knn_model
```

```python
def predict_plant(knn_model, symptoms_vector, n_recommendations=5):
    res= knn_model.predict(symptoms_vector)
    return res
```

```python
def main():
    no_of_symptoms = int(input("Enter the number of symptoms: "))
    symptoms_arr = []
    for i in range(no_of_symptoms):
        symptom = input(f"Enter symptom {i+1}: ")
        symptoms_arr.append(symptom)
    symptoms_vector = vectorize_input(symptoms_arr)
    knn_model = train_knn_model()
    recommended_plants = predict_plant(knn_model, [symptoms_vector], n_recommendations=5)
    print(f"\nPlant recommendations for {symptoms_arr[0:]}")
    for plant in recommended_plants:
        print(f"- {plant}")
```

```python
main()
```
